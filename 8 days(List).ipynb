## 6. wap reverse the list using two pointer?

```python
def list_reverse(l):
   i,j=0,len(l)-1
   while i<j:
       l[i],l[j]=l[j],l[i]
       i+=1
       j-=1
   return l
# l=eval(input("Enter a list: "))
l=[10, 20, 30, 40, 44, 50, 55,60]
print(list_reverse(l))
```

## 7. Wap to check whether the list is palindrome or not using two pointer?

```python
#using 2 poniters
def is_palindrome(l):
    i=0
    j=len(l)-1
    while i<j:
        if l[i]==l[j]:
            i+=1
            j-=1
        else:
            return "Not a palindrome"
    return "Palindrome"
# l=eval(input("Enter a list: "))
l=[10,20,20,10]
print(is_palindrome(l))
print(is_palindrome("AMMA"))
```

```python
def palindrome(l):
    i=0
    j=len(l)-1
    flag=True
    while i<j:
        if l[i]!=l[j]:
            flag=False
            break
        i+=1
        j-=1
    if flag:
        return "Palindrome"
    else:
        return "not a palindrome"
l=eval(input("Enter a list: "))
print(palindrome(l))
```

## 10. wap to rotate the list for n number of items to right or left?

```python
#right rotate
def right_rotate(l,n):
    n=n%len(l)
    out=l[-n:]+l[:-n]
    return out
l=eval(input("Enter a list: ")) #[10,20,5,7,30]
n=int(input("Enter a number of rotations: ")) #3
print(right_rotate(l,n)) #[5, 7, 30, 10, 20]
```

```python
#left rotate
def left_rotate(l,n):
    n=n%len(l)
    out=l[n:]+l[:n]
    return out
l=eval(input("Enter a list: ")) #[10,20,5,7,30]
n=int(input("Enter a number of rotations: ")) #3
print(left_rotate(l,n)) #[7, 30, 10, 20, 5]
```

## 11. Wap to move zeros to end of the list?

```python
def move_zero(l):
    zero=[]
    out=[]
    for i in l:
        if i!=0:
            out.append(i)
        else:
            zero.append(i)
    return out+zero
l=eval(input("Enter a list: ")) #[10,20,0,0,2,3,5,0]
print(move_zero(l))
```

```python
def move_zero(nums):
    if len(nums)==1:
        return nums
    i=0
    while i<len(nums):
        if nums[i]==0:
            break
        i+=1
    if i==len(nums):
        return nums

    j=i+1
    while j<len(nums):
        if nums[j]!=0:
            nums[i],nums[j]=nums[j],nums[i]
            i+=1
        j+=1
    return nums

nums=[10,20,0,0,0,2,3,5,0]

print(move_zero(nums))
```

## 13. Wap to find the pair with given sum?

```python
def pair_sum(l,n):
    out=[]
    for i in range(0,len(l)):
        if l[i]==n:
            out.append(l[i])
        else:
            for j in range(i+1,len(l)):
                if l[i]+l[j]==n:
                    out.append([l[i],l[j]])
    return out
l=eval(input("Enter a list: ")) #[5, 7, 30, 10, 20]
n=int(input("Enter a val: ")) #30
print(pair_sum(l,n)) #[30, [10, 20]]
```

## 20.Wap to find duplicate elements from the list?


```python
def duplicate_num(l):
    original = []
    dup = []
    for i in l:
        if i in original:
            if i not in dup:
                dup.append(i)
        else:
            original.append(i)
    return dup

l = [10,20,30,10,20,20,20]
print(duplicate_num(l))

#or
```

```python
def find_duplicate(l):
    dup = []
    for i in l:
        if l.count(i) > 1 and i not in dup:
            dup.append(i)
    return dup

l = eval(input("Enter a list: ")) #[10,20,30,10,20,20,20]

print("Duplicate elements:", find_duplicate(l))
```

## 23. Wap to find the element which is repated more number of times?


```python
def most_repeated(l):
    freq = {}
    for i in l:
        if i not in freq:
            freq[i] = 1
        else:
            freq[i] += 1
    max_count = 0
    element = None
    for key, value in freq.items():
        if value > max_count:
            max_count = value
            element = key
    return element
l = [10,20,22,30,10,20,40,10]

print(most_repeated(l))
#or
```

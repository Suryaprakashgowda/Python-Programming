## 1. Linear Search:-
```
It is a searching algorithm which works on the priciple called sequential searching priciple.
```
* It will traverse through the collection to find the targeted element.
```
a=[5,7,3,9,8]
if found index if not found -1
```



````python

#Linear Search
def linear_search(col,key):
    for i in range(len(col)):
        if col[i] == key:
            return i
    return -1
col=[1,2,3,4,5,6,7,8,9]
key=5
# col=eval(input("Enter the collection list: ")) #[10,20,30,40,50]
# key=int(input("Enter the key to find: ")) #30
print(linear_search(col,key)) #2 index, 4

#time_complexity:- O(n)
#space_complexity:- O(1)
````

* It is a searching algorithm which works on sorted collection by repeatedly dividing the collection into two halves until we found targeted element.

````python
def binary_search(col,key):
    li=0
    hi=len(col)-1
    while li<=hi:
        mid=(li+hi)//2 #(0+5)//2=2, (3+5)=4,(3+3)//2=3
        if col[mid]==key: #(8==12)❌ , (16==12)❌,(12==12)✅
            return mid
        elif col[mid]<key: #(8<12)✅, (16<12)❌
            li=mid+1
        else:
            hi=mid-1 #(hi= 4-1)
    return -1
col=[5,7,8,12,16,19]
key=12
# col=eval(input("Enter a list: "))
# key=int(input("Enter a key: "))
print(binary_search(col,key))
````

## 1. Bubble sort
```
It is the sorting algorithm where each and every element will get swapped based on condtion in order to sort the values.
```
* It performs n-1 passes to arrange the values.
* In each and every pass, the sorted element will be at the last.
### Steps:-
1. Consider the collection
2. Consider starting value.
3. Compare current value with next value.
4. If current value is greater than next value, then swap it or else keep as it is.
5. Consider the next pair of elements and repeat step 3 and step 4 until the value is sorted
6. After one complete pass,the largest element will be at the last.
7. Repeat the same steps for remaining values by ignoring the last sorted one.

* Time complexity:- O(n*m) -> O(n^2)
* Space Complexity:- O(1)

````python
# def bubble_sort(col):
#     for i in range(0,len(col)-1):
#         for j in range(0,len(col)-1-i):
#             if col[j]>col[j+1]:
#                 col[j],col[j+1]=col[j+1],col[j]
#     return col
# col=[5,3,7,1,3,2,5,12,9,22,32,12,8,87]
# print(bubble_sort(col))

#or
def bubble_sort(col):
    for passno in range(1,len(col)):
        swapped = False
        for j in range(0,len(col)-passno):
            if col[j] > col[j+1]:
                col[j],col[j+1]=col[j+1],col[j]
                swapped = True
        if not swapped: #if not true ,it will loops continues, if not false true break loops
            break
    return col
col=[5,2,1,4,3,7,8,12,16,19]
# col=eval(input("Enter a list: "))
print(bubble_sort(col))

````

## Merge Sort:-
```
Merge Sort uses Divide and Conquer. It recursively divides the list into halves, sorts them, and then merges them back in sorted order.
```
* It works on the principle called Divide and Conquer.
* Will divide the collection into two halves (left and right) continuously.

#### Steps of Algorithm

1. Divide: Divide the collection into two halves, until we get one as a element.
2. Conquer: Recursively sort the left and right collection.
3. Merge: Merge the two sorted collection by comparing the values.


````python
def mergeSort(l):
    if len(l)>1:
        left=l[:len(l)//2]
        right=l[len(l)//2:]
        mergeSort(left)
        mergeSort(right)
        i=0
        j=0
        k=0
        while i<len(left) and j<len(right):
            if left[i]<right[j]:
                l[k]=left[i]
                i+=1
                k+=1
            else:
                l[k]=right[j]
                j+=1
                k+=1
        while i<len(left):
            l[k]=left[i]
            i+=1
            k+=1
        while j<len(right):
            l[k]=right[j]
            j+=1
            k+=1
    return l
# l=eval(input("Enter the values: "))
l=[2,1,4,5,3,7,8,9,6]
mergeSort(l)
print(l)
````

## 1.Stack:- (FILO/LIFO)(FIRST IN LAST OUT / LAST IN FIRST OUT)
* It is the temporary memory which is used to store the value for temporary purpose.
* It works on the principle (FILO) FIRST IN LAST OUT / LAST IN FIRST OUT (LIFO).
* Top is the end where insertion and delection operation takes place.
* ### 2 operations:
* 1. push :- Inserting the value.
  2. pop :- deleting the value.

#### 2 ways to implement stack.
1. Using python list
2. Using inbuilt modules.



## 1. Using list

````python
class stack:
    def __init__(self,size=5):
        self.size=size
        self.st=[]
    def push(self,data):
        if len(self.st)==self.size:
            raise ValueError("Stack is full")
        else:
            self.st.append(data)
    def pop(self):
        if len(self.st)==0:
            raise ValueError("Stack is empty")
        else:
            # return self.st.pop()
            self.st.pop()
    def peek(self):
        # if len(self.st)==0:
        #     raise ValueError("Stack is empty")
        # else:
            return self.st[-1]
    def display(self):
        return self.st

s1=stack()
# s1.pop() #error
s1.push(10)
s1.push(20)
s1.push(30)
s1.push(40)
s1.push(50)
s1.pop()
s1.push(60)
print(s1.display())
print(s1.peek())
# print(s1.pop())
print(s1.display())
print(s1.peek())
````

* It is the temporary memory which is used to store the data for temporary purpose.
* works on the principle called FIRST IN FIRST OUT (FIFO) / LAST IN LAST OUT (LIFO)

### 2 OPERATION

2. POP /DEQUEUE

#### IMPLEMENT QUEUE IN 2 WAYS.
1. USING LIST
2. USING INBUILT MODULE

## 1. Using list

````python
# 1. Using List
class Queue:
    def __init__(self,size=5):
        self.size=size
        self.q=[]
    def push(self,data):
        if len(self.q)==self.size:
            # print("Queue is full")
            raise ValueError("Queue is full")
        else:
            self.q.append(data)
    def pop(self):
        if len(self.q)==0:
            raise ValueError("Queue is empty")
        else:
            self.q.pop(0)
    def peek(self):
        return self.q[0]
    def display(self):
        return self.q

q=Queue()
# q.pop()
q.push(10)
q.push(20)
q.push(30)
q.push(40)
q.push(50)
# q.push(60)
q.pop()
print(q.display())
print(q.peek())
````

## 2.Check Palindrome?


```python
def check_pal(s):
    i=0
    j=len(s)-1
    while i<j:
        if s[i]==s[j]:
            i+=1
            j-=1
        else:
            return "Not Palindrome"
    return "Palindrome"
# s=list(input("Enter a string: ")) #amma #malayalam
s="malayalam"
print(check_pal(s)) #malayalam
```

## 3.Wap to print the count of Vowels and consonants in a given string?


```python
def vowels(s):
    s=s.lower()
    vow=["a","e","i","o","u"]
    vow_count=0
    not_vow_count=0
    for i in range(len(s)):
        if s[i] in vow:
            vow_count+=1
        else:
            not_vow_count+=1
    return vow_count,not_vow_count
# s=input("Enter string: ")#surya
s="surya"
vow_count,not_vow_count=vowels(s)
print(f"vowels:{vow_count},consonants:{not_vow_count}") #vowels:2,consonants:3
```

## 7. Wap to find the duplicate characters from given string?

```python
def duplicate(s):
    ori=""
    dup=""
    for i in s:
        if i not in ori:
            ori+=i
        else:
            dup+=i
    return dup
#s=input("enter a string")
s="suryasss"
print(duplicate(s))

#or

def duplicate_char(s):
    dup=set()
    res=set()
    for i in s:
        if i in res:
            dup.add(i)
        else:
            res.add(i) #h,he,hel,helo
    # return "".join(dup)
    return dup
s="hellolo"
print(duplicate_char(s))

#or
def find_duplicates(s):
    str=""
    dup=""
    for i in s:
        if i not in str:
            str+=i
        elif i not in dup:
            dup+=i
    return dup
s="suryaaasss"
# print(list(find_duplicates(s)))
print(find_duplicates(s))
```

## 9. Wap to Check whether given string are anagram or not?
The string should contain same characters in any order by ignoring spaces and cases


```python
#s1="Cat"
#s2="act"

def ananagram(s1,s2):
    if sorted(s1)==sorted(s2):
        return "Anagram"
    else:
        return "Not anagram"

# s1=input("Enter first string:").replace(" ","").lower()
# s2=input("Enter second string:").replace(" ","").lower()
s1="Cat is Dog"
s2="Act is dog"
s1=s1.lower().replace(" ","")
s2=s2.lower().replace(" ","")
print(ananagram(s1,s2))
```

## 10. Wap to Check whether string is panagram or not?
if the string consist of all the alphabetical character from A to Z.
Ex:- "The quick brown fox jumps over the lazy dog"


```python
import string
def panagram(s):
    str_set=set(s)
    # print(str_set)
    alpha=set(string.ascii_lowercase) #storing all the lower aplhabits
    # print(alpha)
    if alpha.issubset(str_set):
        return "Panagram"
    else:
        return "Not Panagram"
# s=input("enter a string").lower()
s="The quick brown fox jumps over the Lazy dog".lower()
# s=s.lower()
print(panagram(s))
```

## 18.WAP to Count Uppercase, Lowercase, Digits, Special Symbols & Spaces?

```python
# s = input("Enter a string: ")
s="How are you 1 2 * ^"
upper = lower = digit = special = space = 0

for i in s:
    if i.isupper():
        upper += 1
    elif i.islower():
        lower += 1
    elif i.isdigit():
        digit += 1
    elif i == " ":
        space += 1
    else:
        special += 1
print("Uppercase:", upper)
print("Lowercase:", lower)
print("Digits:", digit)
print("Spaces:", space)
print("Special Symbols:", special)
```

## 23.Wap to print largest palindrome substring from given string?
##### s="abacd'
a=s[0:1], a==a ✅ then largest len add it.
##### ab=s[:2], ab!=ba then avoid
##### aba=s[:3] aba==aba then largest update, until [5:5] it goes on

```python
def largest_substring(s):
    largest=""
    for i in range(len(s)):
        for j in range(i+1,len(s)+1):
            sub=s[i:j]
            if sub==sub[::-1] and len(sub)>len(largest):
                largest=sub
    return largest
# s=input("Enter a string: ") #abacdaaabbaaadc, #abacd
# s="abacdaaabbaaadc"
s="abacd"
print(largest_substring(s)) #cdaaabbaaadc, #aba
```

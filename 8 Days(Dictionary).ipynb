## 1. wap to count the frequency of character from the string using dictionary.

````python
def dict_Freq(s):
    s1={}
    for i in s:
        if i not in s1:
            s1[i]=1
        else:
            s1[i]+=1
    return s1
# s=input("Enter a string")
s="Hello"
print(dict_Freq(s))

#simple
#{'H': 1, 'e': 1, 'l': 2, 'o': 1}
#var[key]=new_val
#var.count(var)
s="Hello"
out={}
for i in s:
    out[i]=s.count(i)
print(out)
````

## 2. Wap to merge two dictionary.
```
d1={'a':10,'b':'20}
d2={'c':30,'d':40}
{'a':10,'b':'20,'c':30,'d':40}
update() to merge dict
var.upate(var2)
```

````python
def merge_dict(d1,d2):
    out=d1
    for i in d2:
        if i not in out:
            out[i]=d2[i]
        else:
            out[i]=d2[i]
            # out[i]=out[i]+d2[i]
    return out
d1={'a':10,'b':20,'c':10}
d2={'c':30,'d':40}
print(merge_dict(d1,d2))

#OR
d1={'a':10,'b':20,'c':10}
d2={'c':30,'d':40}
d1.update(d2)
print(d1)

#or
d1={'a':10,'b':20,'c':10}
d2={'c':30,'d':40}
out=d1
for i in d2:
    out[i]=d2[i]
print(out)
````

## 3. Wap to sort the dictionary by keys?


````python

# d=eval(input("Enter a dictionary"))
d={"x":5,'a':10,'m':30}
key=list(d.keys())
for passno in range(1,len(key)):
    for i in range(len(key)-passno):
        if key[i]>key[i+1]:
            key[i],key[i+1]=key[i+1],key[i]
out={}
for i in key:
    out[i]=d[i]
print(out) #{'a': 10, 'm': 30, 'x': 5}
````

## 7. Wap to remove duplicates from the dictionary?


````python
# d=eval(input("Enter a dictionary"))
d={'a':10,'b':20,'c':10,'d':30}
key_val={}
res=set()
for k,val in d.items():
    if val not in res:
        key_val[k]=val
        res.add(val)
print(key_val)
````

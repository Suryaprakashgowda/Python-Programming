## 1. wap to count the frequency of character from the string using dictionary.

````python
def dict_Freq(s):
    s1={}
    for i in s:
        if i not in s1:
            s1[i]=1
        else:
            s1[i]+=1
    return s1
# s=input("Enter a string")
s="Hello"
print(dict_Freq(s))

#simple
#{'H': 1, 'e': 1, 'l': 2, 'o': 1}
#var[key]=new_val
#var.count(var)
s="Hello"
out={}
for i in s:
    out[i]=s.count(i)
print(out)
````

## 2. Wap to merge two dictionary.
```
d1={'a':10,'b':'20}
```

````python
def merge_dict(d1,d2):
    out=d1
    for i in d2:
        if i not in out:
            out[i]=d2[i]
        else:
            out[i]=d2[i]
            # out[i]=out[i]+d2[i]
    return out
d1={'a':10,'b':20,'c':10}
d2={'c':30,'d':40}
print(merge_dict(d1,d2))

#OR
d1={'a':10,'b':20,'c':10}
d2={'c':30,'d':40}
d1.update(d2)
print(d1)

#or
d1={'a':10,'b':20,'c':10}
d2={'c':30,'d':40}
out=d1
for i in d2:
    out[i]=d2[i]
print(out)
````

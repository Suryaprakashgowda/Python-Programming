## Time Complexity
> It is total time taken for execution

>> # 3 ways:
>>>  1. Best time complexity
>>> 2. worst time complexity
>>> 3. Average time complexity

## 1.Best time complexity( Ω )(omega)
```
a=[10,20,30,40]
find a=10 takes less time .
```
```
It is the measures of minimum time taken by the program get postive respone
```
## 2. Worst time complexity(Big O notation (O))
```
a=[10,20,30,40]
find a=40 takes more time.
```
```
It is the measure of maximum time taken by the program to get postive or negative response.
```
## 3. Average time complexity (Θ) (theta)
```
a=[10,20,30,40,50]
find a=30 takes avg time.
```
```
It is the measure of average time taken by the program to get postive respone.
```

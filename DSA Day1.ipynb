## Searching algorithm
```
It is the algorithm which is used to search whether the value is present in collection or not.
```
* If the value is present, we will return the index of the value.
* if value is not present, we return -1.
### 3 Types:
> 1. Linear Search
> 2. Binary Search
> 3. Ternary Search

## 1. Linear Search:-
```
It is a searching algorithm which works on the priciple called sequential searching priciple.
```
* It will traverse through the collection to find the targeted element.
```
a=[5,7,3,9,8]
if found index if not found -1
```



````python

#Linear Search
def linear_search(col,key):
    for i in range(len(col)):
        if col[i] == key:
            return i
    return -1
col=[1,2,3,4,5,6,7,8,9]
key=5
# col=eval(input("Enter the collection list: ")) #[10,20,30,40,50]
# key=int(input("Enter the key to find: ")) #30
print(linear_search(col,key)) #2 index, 4

#time_complexity:- O(n)
#space_complexity:- O(1)
````

## 2. Binary Search
* It is a searching algorithm which works on sorted collection by repeatedly dividing the collection into two halves until we found targeted element.
```
key=12
    0 1  2  3  4
l=[10,12,18,21,28] (sorted list)
    Li         Lj
while li<=lj:
    mid=(Li + Lj)//2
        (0+4)//2 =>2
    if l[mid]==key:
        return mid
    elif l[mid]<key: (18<12)❌
        li=mid+1
    elif l[mid]>key:
        lj=mid-1
return -1

Time complexity: O(logn)
explain:-
n/2 = n/2^1
n/4 = n/2^2
n/8 = n/2^3
..
n/n = n/2^k
n/2^k=1 (apply both  log)
n=2^k
log(n) =log(2^k)
log(n)=1 (log(2^k) is 1) (proved why O(logn)

 Space complexity: O(1)
```


````python
def binary_search(col,key):
    li=0
    hi=len(col)-1
    while li<=hi:
        mid=(li+hi)//2 #(0+5)//2=2, (3+5)=4,(3+3)//2=3
        if col[mid]==key: #(8==12)❌ , (16==12)❌,(12==12)✅
            return mid
        elif col[mid]<key: #(8<12)✅, (16<12)❌
            li=mid+1
        else:
            hi=mid-1 #(hi= 4-1)
    return -1
col=[5,7,8,12,16,19]
key=12
# col=eval(input("Enter a list: "))
# key=int(input("Enter a key: "))
print(binary_search(col,key))
````

## 1. Bubble sort
```
It is the sorting algorithm where each and every element will get swapped based on condtion in order to sort the values.
```
* It performs n-1 passes to arrange the values.
* In each and every pass, the sorted element will be at the last.
### Steps:-
1. Consider the collection
2. Consider starting value.
3. Compare current value with next value.
4. If current value is greater than next value, then swap it or else keep as it is.
5. Consider the next pair of elements and repeat step 3 and step 4 until the value is sorted
6. After one complete pass,the largest element will be at the last.
7. Repeat the same steps for remaining values by ignoring the last sorted one.

* Time complexity:- O(n*m) -> O(n^2)
* Space Complexity:- O(1)

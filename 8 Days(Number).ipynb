## 1. Wap to check Prime number or not?
int(n**0.5)+1

👉 checks only till square root
👉 faster solution
```
We check divisibility only up to √n because if a number has a factor larger than √n, the corresponding factor must be smaller than √n.
This reduces time complexity from O(n) to O(√n).
```

````python
def prime(n):
    if n <= 1:
        return "Not Prime"
    for i in range(2, n):
        if n % i == 0:
            return "Not Prime"
    return "Prime Number"

# n = int(input("Enter number: "))
n=7
print(prime(n))

#or

def prime_num(n):
    count=0
    for i in range(1,n+1):
        if n%i==0:
            count+=1
    if count==2:
        return "Prime number"
    else:
        return "Not Prime Number"
# n=int(input("Enter the number"))
n=10
print(prime_num(n))
````

```
🧠 What is n**0.5?
n**0.5
means:
√n
(square root of n)

Example:
n = 36
n**0.5 = 6
🧠 Why Check Only Till √n?

Example:
n = 36

Factors:
1 × 36
2 × 18
3 × 12
4 × 9
6 × 6

After 6 (√36), factors start repeating:
9 × 4
12 × 3
18 × 2
36 × 1

So if a number has a factor, one of its factors must be ≤ √n.
Therefore, we only need to check till √n.

Your Code
Time Complexity = O(√n)
Space Complexity = O(1)
```

````python
def prime(n):
    if n<2:
        return "Not Prime"
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return "Not Prime"
    return "Prime"
print(prime(23))
````

## 4.Wap to check whether the given number is armstrong number or not?
```
Armstrong number:- Sum of power of individual digit == given number
n=12
  1^2+2^2=5 (not a armstrong number)
n=153
   1^3+5^3+3^3=153(Armstrong number)
```

````python
def armstrong(n):
    sum=0
    for i in str(n):
        sum=sum+int(i)**len(str(n))
    return "Armstrong" if sum==n else "Not Armstrong"
print(armstrong(153))

#or
def armstrong_num(n):
    sum=0
    for i in str(n):
        num=int(i)
        sum=sum+num**len(str(n))
        # print(sum)
    if sum==n:
        return "Armstrong"
    else:
        return "Not Armstrong"
n=int(input("Enter Number:")) #153, 123-> not a armstrong number
print(armstrong_num(n))
````

## 5.Wap to check whether the given number is strong number or not?
#### Strong number:-
* Sum of factorial of individual digit == given number
```
n=12
   1!+2!
   1+2*1=3 (not strong number)
n=145
   1!+4!+5!
    1+(4*3*2*1)+(5*4*3*2*1)=145 (Strong number)
```

````python
def strong_num(n):
    sum=0
    for i in str(n):
        num=int(i)
        fact=1
        for j in range(1,num+1):
            fact=fact*j
        sum=sum+fact
    if sum==n:
        return "Strong Number"
    else:
        return "Not Strong Number"
n=int(input("Enter Number:")) #145-> strong number ,#12 ->not a strong number
print(strong_num(n))
````

## 6. Wap to check whether given number is perfect number or not?
#### Perfect number:-
* sum of divisor/factors== given number
```
n=8
  1+2+4=7 (we need to exclude 8 , sum all the divisor value it should be equal to given number then we call it
as perfect number if not not perfect number)
n=6
  1+2+3=6 (perfect number)
```

````python
def perfect_number(n):
    sum=0
    for i in range(1,n): #(1,2,3,4,5) (6 excluded)
        if n%i==0:
            sum=sum+i
    if sum==n:
        return "Perfect Number"
    else:
        return "Not Perfect Number"
n=int(input("Enter Number:")) #6
print(perfect_number(n)) #perfect number
````

## 8. Wap to find GCD / HCF of two number?
HCF (Highest Common Factor) and GCD (Greatest Common Divisor)
```
“GCD is the greatest number that divides both numbers completely.”
> GCD(a,b) = GCD(b, a%b)
✅ Example
Input:
    a = 12
    b = 18
Factors:
    12 → 1,2,3,4,6,12
    18 → 1,2,3,6,9,18
Common highest:
    6
Output:
    GCD = 6
```

````python
def gcd(a, b):
    small = min(a, b)
    for i in range(small, 0, -1):
        if a % i == 0 and b % i == 0:
            return i
# a = int(input("Enter first number: "))
a=12
# b = int(input("Enter second number: "))
b=18
print("GCD =", gcd(a, b))

#or
# Using Euclidean Algorithm (Best for Interview )
# a=12, b=18
#
# a=18, b=12
# a=12, b=6
# a=6, b=0
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a
print(gcd(12,18))
````

## 9. Wap to find LCM of two number?
```
“LCM is the smallest number divisible by both given numbers.”
LCM × GCD = a × b
✅ Example

Input:
    a = 4
    b = 6
Multiples:
    4 → 4,8,12,16...
    6 → 6,12,18...
First common multiple:
    12
Output:
    LCM = 12
```

````python
def lcm(a, b):
    greater = max(a, b)
    while True:
        if greater % a == 0 and greater % b == 0:
            return greater
        greater += 1
        # print(greater)
# a = int(input("Enter first number: "))
# b = int(input("Enter second number: "))
a=4
b=6
print("LCM =", lcm(a, b))

#or
# a = 4
# b = 6
#
# GCD = 2
#
# LCM = (4×6)//2
# LCM = 24//2
# LCM = 12

def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    # print(a)
    return a
def lcm(a, b):
    return (a * b) // gcd(a, b)

print("Lcm:",lcm(4,6))
````

## 10. Wap to convert decimal to binary?

```

## 11.Wap to convert binary to decimal?
```

✅ Step 3

Total:
    8 + 2 = 10
```

Time Complexity

```

## 12. Wap to swap two numbers with using temporary variable?
```
"Store one value in a temporary variable and interchange the values.”
Initial:
a = 10
b = 20
✅ Step 1
    temp = a
    temp = 10
✅ Step 2
    a = b
    a = 20
✅ Step 3
    b = temp
    b = 10
✅ Final Values
    a = 20
    b = 10
🎯 Time Complexity
    O(1)
🎯 Space Complexity
    O(1)
Because:
👉 only one extra variable used.
```


````python
def swap_temp(a,b):
    temp=a
    a=b
    b=temp
    return a,b
# a=int(input("Enter Number:"))
# b=int(input("Enter Number:"))
a=10
b=20
a,b=swap_temp(a,b)
print("a =", a)
print("b =", b)
````

## 1. Wap to check Prime number or not?
int(n**0.5)+1

👉 checks only till square root
👉 faster solution

```python
def prime(n):
    if n <= 1:
        return "Not Prime"
    for i in range(2, n):
        if n % i == 0:
            return "Not Prime"
    return "Prime Number"

# n = int(input("Enter number: "))
n=7
print(prime(n))

#or

def prime_num(n):
    count=0
    for i in range(1,n+1):
        if n%i==0:
            count+=1
    if count==2:
        return "Prime number"
    else:
        return "Not Prime Number"
# n=int(input("Enter the number"))

print(prime_num(n))
```

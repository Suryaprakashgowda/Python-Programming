## Primary Diagonal;-

````python
n=5
for i in range(1,n+1):
    for j in range(1,n+1):
        if i==j: #primary diagonal
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
````

````python
n=5
for i in range(1,n+1):
    for j in range(1,n+1):
        if i<j: #upper triangle
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
````

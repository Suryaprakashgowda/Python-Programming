## Regular Expression.
```
It is the phenomenon of writting an expression to match the pattern inorder to fetch the required data from huge data.

Expression:-
------------
[A-Z] -> It is used to match only one character from A to Z

[a-z] -> It is used to match only one charcter from a to z.

[0-9] or \d  -> It is used to match only one digit from 0 to 9

[A-Z a-z 0-9] or \w -> It is used to macth only one character between A to Z or a to z or 0-9.

[A-Z]{n} -> To match 'n' number of character b/w A to Z.

[A-Z]{m,n} -> To match m,n number of character b/w A-Z.
example :- in -> m=2 and  'and' -> n=3

+ -> used to match 1 to n number of character.
* -> used to match 0 to n number of character.
? -> used to match either 0 or 1 character.

Remove special character properties we use '\'
\* -> normal char
```

## import re

* ### findall() :- used to get matched data from huge data
* var=re.findall('pattern','data')

* ### finditer():-  used to get index of matched data
* var=re.finditer('pattern','data')
* print(var) -> address of obj
* print(list(var)) -> list output


````python
from re import finditer

s = '''In the bustling city of New York, on the chilly morning
of at precisely 07:45 AM, Michael Johnson placed an online
order for a brand-new Samsung Galaxy S22. The order confirmation
email (Order ID: ORD-2023-9876) was sent to his registered email,
michael.johnson@example.com and message sent to registered
phone number 9876534008. He paid $799.99 using his Visa card
ending in 3456, and the transaction was processed 1234567890 79809876955677
by Stripe on 12/12/2023 at 07:50 AM'''

import re
print(re.findall(r'\d{10}', s))
out=re.findall('[6-9][0-9]{9}', s)
print(out)
res=re.finditer('[6-9][0-9]{9}', s)
print(res) #address of object
print(list(res)) # list output
````

## 1. Wap to factorial of given number?


````
def factorial(n):
    if n==0 or n==1:
        return 1
    # else:
    #     return n*factorial(n-1)
    return n*factorial(n-1)
print(factorial(5))
````

## 2. Wap to find the sume of n natural number?

````
def sum_natural(n):
    if n==1:
        return 1
    return n+sum_natural(n-1)
print(sum_natural(4))
````

```
🧠 What is @lru_cache in Python?

@lru_cache is a decorator from the functools module used to cache (remember) function results.

LRU = Least Recently Used

->Why Use It?
Without cache:

def fib(n)
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)

For:
fib(50)
Many values are calculated repeatedly.

With @lru_cache
from functools import lru_cache

@lru_cache(maxsize=None)
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)

print(fib(50))

Python stores previous results.

How It Works
Without Cache
fib(5)

├── fib(4)
│   ├── fib(3)
│   │   ├── fib(2)
│   │   └── fib(1)
│   └── fib(2)
└── fib(3)

Notice:

fib(3)
fib(2)

are calculated multiple times.

With Cache

->First time:
fib(3)
computed and stored.
->Next time:
fib(3)

retrieved from cache.

No recalculation.
```

## Steps to convert looping program into recursion.
1. Initialise the looping variable inside formal arguments section.
2. Termination condtion should be written exactly opposite to looping condtion in the form of if constion.
3. Return the total result inside the termination condtion.
4. Logic of the program should be written as it is.
5. Update the looping variable inside recursive call.

## 3. Wap to extract Lowercase character from the given str?


````
def extract_lowercase(s,res="",i=0):
    if i>=len(s):
        return res
    if 'a'<=s[i]<='z':
        res=res+s[i]
    return extract_lowercase(s,res,i+1)
# s=input("Enter a string:") #Hello
# print(extract_lowercase(s)) #ello
print(extract_lowercase("Abc"))
````

## 4. Wap to get given output?
* input:- [10,'data',4.5,'hai',(1+2j)]
* output:- [10,'ataddata',4.5,'iahhai',(1+2j)]


````
def sample(l,out=[],i=0):
    if i>=len(l):
        return out
    if type(l[i])==str:
        out.append(l[i][::-1]+l[i])
    else:
        out.append(l[i])
    return sample(l,out,i+1)
l=[10,'data',4.5,'hai',(1+2j)]
print(sample(l))
````

## 5. Wap to get following output.?
* l=[5,4.5,'data',[3.2,8],12]
* output:- 25

````
def sample_out(l,sum=0,i=0):
    if i>=len(l):
        return sum
    if type(l[i])==int:
        sum=sum+l[i]
    elif type(l[i])==list:
        sum=sum+sample_out(l[i])
    return sample_out(l,sum,i+1)
l=[5,4.5,'data',[3.2,8],12]
# l=eval(input("Enter a list: "))
print(sample_out(l))
````
